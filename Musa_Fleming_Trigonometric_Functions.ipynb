```python
import numpy as np
import matplotlib.pyplot as plt

function = input("Enter Cos, Tan or Sine: ")

x = np.linspace(-np.pi, np.pi, 100)

if function == "cos":
  y = np.cos(x)
  plt.plot(x, y)
  plt.title("Cos Function")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.grid(True)
  plt.axhline(0, color="black", lw=1)
  plt.axvline(0, color="black", lw=1)

elif function == "tan":
  a = np.linspace(-5 * np.pi, 5 * np.pi, 1000)
  b_tan = np.tan(a)
  b_tan[:-1][np.diff(b_tan) < 0] = np.nan #this line of code removes the lines connecting the different lines.
  plt.plot(a, b_tan)
  plt.ylim(-10, 10)
  plt.title("Tan Title")
  plt.grid(True)
  plt.axhline(0, color="black", lw=1)
  plt.axvline(0, color="black", lw=1)

elif function == "sine":
  y = np.sin(x)
  plt.plot(x, y)
  plt.title("Sine Function")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.grid(True)
  plt.axhline(0, color="black", lw=1)
  plt.axvline(0, color="black", lw=1)

else:
  print("Please enter one of the options named")
```
