```python
import numpy as np
import matplotlib.pyplot as plt

function = input("Enter Cos, Tan or Sine: ")

x = np.linspace(-np.pi, np.pi, 100)

if function.lower() == "cos":
  y = np.cos(x)

elif function.lower() == "tan":
  y = np.tan(x)

elif function.lower() == "sine":
  y = np.sin(x)

else:
  print("Please enter one of the options named")

plt.plot(x, y)
plt.title("Trigonometric Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.axhline(0, color="black", lw=1)
plt.axvline(0, color="black", lw=1)
plt.show
```
